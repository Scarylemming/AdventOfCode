```python
import numpy as np
import math
import matplotlib.pyplot as plt
import copy
f = open("9.txt", "r")


data = [line.split(" ") for line in f.read().split("\n")]
if data[-1] in ["", [""]] : 
    data = data[:-1]
```

```python
head = np.array([0,0])
tail = np.array([0,0])

t_visited = []

def move_h(move, head = head) : 
    if move == "U" : 
        return np.array([0,-1]) 
    elif move == "D" : 
        return np.array([0,1])
    elif move == "R" :
        return np.array([1,0])
    elif move == "L" :
        return np.array([-1,0])

def move_l(head = head, tail = tail) : 
    if abs(head[0] - tail[0]) == 2 : 
        tail[1] = head[1]
        tail[0] += (head[0] - tail[0]) / 2
    elif abs(head[1] - tail[1]) == 2 :
        tail[0] = head[0]
        tail[1] += (head[1] - tail[1]) / 2
    else : 
        pass #Il n'y rien Ã  faire ici
    return tail
```

```python
cases = []
head = np.array([0,0])
tail = np.array([0,0])

for i in data : 
    #try : 
    for j in range(int(i[1])) : 
        head += move_h(i[0])
        tail = move_l(head, tail)
        #print(tail)
        if [tail[0], tail[1]] not in cases : 
            cases.append(copy.copy([tail[0], tail[1]]))
        else : 
            pass
            #print(cases, tail)
    #except :
    #   print(i)
len(cases)
```
