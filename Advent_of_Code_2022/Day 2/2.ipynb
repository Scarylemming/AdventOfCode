```python
f = open("2.txt", "r")


data = [line for line in f.read().split("\n")][:-1]

#data = ["A Y", "B X", "C Z"]
score = 0
for i in data : 
    for k in [-1] : 
        if i[k] in ["A", "X"] : 
            score += 1
        elif i[k] in ["Y", "B"] : 
            score += 2
        elif i[k] in ["Z", "C"] : 
            score += 3
    if i in ["A X", "B Y", "C Z"] : 
        score += 3 
    elif i in ["A Y", "B Z", "C X"] : 
        score += 6
    #print(score)
print(score)

```
