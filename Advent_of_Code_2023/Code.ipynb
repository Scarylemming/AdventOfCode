```python
def information(line) : 
    res = []
    count = 0
    for i in line : 
        if i != "#" : 
            if count != 0 : 
                res.append(count)
                count = 0
        else :
            count += 1
    if count != 0 :
        res.append(count)
    return res
```

```python
def count_possibilities(line,info) :
    res = 0
    if "?" in line : 
        for i,el in enumerate(line) : 
            if el == "?" :
                res += count_possibilities(line[:i] + "." + line[i+1:],info)
                res += count_possibilities(line[:i] + "#" + line[i+1:],info)
                break
    else :
        if information(line) == info : 
            res += 1
    return res
                
```
